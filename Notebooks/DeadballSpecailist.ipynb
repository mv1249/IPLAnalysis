```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
df.head(3)
```

```python
df = df.drop(['dismissal_kind', 'player_dismissed', 'fielder', 'extras_type'],axis = 1)
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

```python
deadball = df.loc[(df['over']>=15)&(df['over']<=20)&(df['is_wicket']==1)]
deadball.shape
```

```python
deadball.head(2)
```

```python
team_list = ['Kolkata Knight Riders',
 'Royal Challengers Bangalore',
 'Kings XI Punjab',
 'Chennai Super Kings',
 'Rajasthan Royals',
 'Mumbai Indians',
 'Sunrisers Hyderabad',
 'Deccan Chargers',
 'Delhi Daredevils',
 'Delhi Capitals']
```

```python
def
```

```python
def player_count(team_name):
    name_list =  []
    temp = deadball.loc[deadball['bowling_team']==team_name]
    player_count = Counter(temp['bowler'])
    player_count = sorted(player_count.items(),key=lambda x:x[1],reverse=True)
    player_count = player_count[:2]
    for i in range(0,len(player_count)):
        name_list.append(player_count[i][0])
    
    return name_list
```

```python
name,wickets = player_count('Kolkata Knight Riders')
name,wickets
```

```python
deadballers = []
for i in team_list:
    deadballers.append(player_count(i))
    
deadballers
```

```python
len(deadballers)
```

```python
deadballers = [deadballers[i][j] for i in range(10) for j in range(2)]
deadballers
```

```python
def getwicketcount(team_name,bowler_name):
    mapper = deadball.loc[(deadball['batting_team']==team_name) & (deadball['bowler']==bowler_name)]
    return mapper.shape[0]
```

```python
wicket_perteam = []
for i in range(0,len(deadballers)):
    subset = []
    for j in range(0,len(team_list)):
        subset.append(getwicketcount(team_list[j],deadballers[i]))
        
    wicket_perteam.append(subset)
    
wicket_perteam
```

```python
len(deadballers),len(wicket_perteam)
```

```python
dead_mapper = dict(zip(deadballers,wicket_perteam))
dead_mapper
```

```python
dead_df = pd.DataFrame(dead_mapper)
dead_df = dead_df.T
dead_df
```

```python
sum(wicket_perteam[0])
```

```python
dead_df.columns = team_list
dead_df.head()
```

```python
total_wickets = [sum(wicket_perteam[i]) for i in range(0,len(deadballers))]
```

```python
len(total_wickets)
```

```python
dead_df['Total_wickets'] = dead_df.sum(axis = 1)
dead_df
```
