## Last over specialists

```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df_ball_by_ball = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
df_ball_by_ball.head()
```

```python
df_ball_by_ball.isnull().sum()
```

```python
df_ball_by_ball = df_ball_by_ball.drop(['dismissal_kind', 'player_dismissed', 'fielder', 'extras_type'],axis = 1)
df_ball_by_ball = df_ball_by_ball.dropna()
df_ball_by_ball.isnull().sum()
```

```python
df_ball = df_ball_by_ball.copy()
df_ball.head(2)
```

```python
df_deadball = df_ball.loc[(df_ball['over']>=15) & (df_ball['over']<=20)]
df_deadball.head()
```

```python
df_deadball['over'].unique()
```

```python
df_kkr = df_deadball[df_deadball['batting_team']=='Kolkata Knight Riders']
df_rcb = df_deadball[df_deadball['batting_team']=='Royal Challengers Bangalore']
df_kxip = df_deadball[df_deadball['batting_team']=='Kings XI Punjab']
df_csk = df_deadball[df_deadball['batting_team']=='Chennai Super Kings']
df_rr = df_deadball[df_deadball['batting_team']=='Rajasthan Royals']
df_mi = df_deadball[df_deadball['batting_team']=='Mumbai Indians']
df_sun = df_deadball.loc[(df_deadball['batting_team']=='Sunrisers Hyderabad')]
df_dc =  df_deadball.loc[(df_deadball['batting_team']=='Deccan Chargers')]                          
df_dd = df_deadball.loc[(df_deadball['batting_team']=='Delhi Daredevils')]
df_dcap = df_deadball.loc[df_deadball['batting_team']=='Delhi Capitals']
```

```python
hyd_count = Counter(df_dc['batsman'])
hyd_count = sorted(hyd_count.items(),key = lambda x:x[1],reverse=True)
hyd_count[0][:5]
```

```python
df_list = [df_kkr,df_rcb,df_kxip,df_csk,df_rr,df_mi,df_sun,df_dc,df_dd,df_dcap]
```

```python
def getcountplayer(df):
    player = Counter(df['batsman'])
    count = sorted(player.items(),key = lambda x:x[1],reverse = True)
    return count[:5]
```

```python
def playername(mapper):
    name_list = []
    for i in range(0,len(mapper)):
        name_list.append(mapper[i][0])
        
    return name_list
```

```python
main_list = []
for i in df_list:
    #subset = []
    mapper = getcountplayer(i)
    main_list.append(playername(mapper))
```

```python
def getruns(name,team):
    runs = df_deadball.loc[(df_deadball['batsman']==name)&(df_deadball['batting_team']==team)]
    return sum(runs['total_runs'])

```

```python
df_dead_ball = pd.DataFrame(name_list,columns=['Player Name'])
df_dead_ball.head(3)
```

```python
def getfour(name):
    mapper = df_deadball.loc[(df_deadball['batsman']==name)&(df_deadball['total_runs']==4)]
    return mapper.shape[0]
```

```python
def getsix(name):
    mapper = df_deadball.loc[(df_deadball['batsman']==name)&(df_deadball['total_runs']==6)]
    return mapper.shape[0]
```

```python
four_list = []
for i in name_list:
    four_list.append(getfour(i))
    
```

```python
six_list = []
for i in name_list:
    six_list.append(getsix(i))
```

```python
df_dead_ball['Team_name'] = teamlist
df_dead_ball.head()
```
