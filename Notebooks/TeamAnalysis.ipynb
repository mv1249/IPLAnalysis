```python
def teamwin(team_name,year):
    matcher = df_matches.loc[(df_matches['winner']==team_name) & (df_matches['Year']==year)]
    return matcher.shape[0]
```

```python
year_list = [i for i in range(2008,2021)]
year_list
```

```python
teams_list = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Delhi Daredevils',
       'Royal Challengers Bangalore', 'Rajasthan Royals',
       'Kings XI Punjab', 'Deccan Chargers', 'Mumbai Indians',
       'Pune Warriors', 'Kochi Tuskers Kerala',
       'Sunrisers Hyderabad', 'Rising Pune Supergiants', 'Gujarat Lions',
       'Rising Pune Supergiant', 'Delhi Capitals']
```

```python
len(teams_list),len(year_list)
```

```python
totalwins = []
for i in range(0,len(teams_list)):
    susbset = []
    for j in range(0,len(year_list)):
        susbset.append(teamwin(teams_list[i],year_list[j]))
        
    totalwins.append(susbset)
```

```python
totalwins
```

```python
win_dict = dict(zip(teams_list,totalwins))
win_dict
```

```python
df_random = pd.DataFrame(win_dict)
df_random = df_random.T
df_random
```

```python
df_random.columns = year_list
df_random
```

```python
total_sums = [sum(totalwins[i]) for i in range(0,len(totalwins))]
total_sums
```

```python
df_random['Total_wins'] = total_sums
df_random
```

```python
df_random = df_random.sort_values('Total_wins',axis = 0,ascending=False)
df_random
```

```python
df_random.to_csv('TeamWins.csv')
```
