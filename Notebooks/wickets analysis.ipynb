### gives the Count of the bowler

```python
def getcountplayer(df):
    player = Counter(df['bowler'])
    player = sorted(player.items(),key = lambda x:x[1],reverse = True)
    return player[:5]

```

### gives the map of bowler team and name

```python
def formmap(df):
    player = df['bowler']
    team = df['bowling_team']
    bowler_map = dict(zip(player,team))
    
    return bowler_map
```

## Retriving stats of Top players performance against each team

```python
def topbowlerstats(name,team_name):
    player = temp.loc[(temp['bowler']==name) & (temp['is_wicket']==1) & (temp['batting_team']==team_name)]
    return sum(player['is_wicket'])

```

```python
team_list = ['Kolkata Knight Riders','Royal Challengers Bangalore','Kings XI Punjab','Chennai Super Kings',
            'Rajasthan Royals','Mumbai Indians','Sunrisers Hyderabad','Deccan Chargers','Delhi Daredevils','Delhi Capitals']
```
